```python
def table4(n):
  result = n*4
  return result
  print ("resultado")

a = table4(3)
print(a)
```

```python
def table5(n):
  result1 = n*4
  result2 = n*5
  result3 = n*6
  
  yield result1
  yield result2
  yield result3

b = table5(3)
print(b.__next__())
```

**Creating generator**

```python
# --- function ---
def natural_No(n):
  result = []
  for i in range(1,n):
    result.append(i)
  return result

n = natural_No(10)
print(n)
```

```python
# --- Generator ---
def natural2(n):
  for i in range(1,n):
    yield(i)

n1 = natural2(10)
for i in n1:
  print(i)
```
