**Ternary operators**

```python
condition = False
if condition:
  x = 1
else:
  x = 0

print(x)
```

```python
condition = False

x = 1 if condition else 0

print(x)
```

**Enumerate function**

```python
names = ['Corey', 'Chris', 'Dave', 'Travis']

index = 0
for name in names:
  print(index, name)
  index += 1
```

```python
for index, name in enumerate(names):
  print(index, name)
```

**loop over two lists at once**

```python
names = ['Peter Parker', 'Clarck Kent', 'Wade Wilson', 'Bruce Wayne']
heroes = ['Spiderman', 'Superman', 'Deadpool', 'Batman']

for index, name in enumerate(names):
  hero = heroes[index]
  print(f'{name} is actually {hero}')
```

```python
for name, hero in zip(names, heroes): # en dos listas o mas
  print(f'{name} is actually {hero}')
```
