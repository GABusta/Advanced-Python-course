**List Comprehensions**

```python
num = [1, 2, 3, 4, 5]
squares = []

for n in num:
  squares.append(n**2)
print (squares)
```

```python
squares2 = [n**2 for n in num]
print(squares2)
```

Common numbers of lists, using list comprehension

```python
list1 = [1, 2, 3, 4, 5]
list2 = [2, 3, 4, 5, 6]
intersection = []

for x in list1:
  for y in list2:
    if x == y:
      intersection.append(x)

print(intersection)
```

```python
intersec = [x for x in list1 for y in list1 if x == y]
print(intersec)
```

```python
nonCommonElem = [(x, y) for x in list1 for y in list1 if x != y]
print(nonCommonElem)
```

**Lower case all the letters**

```python
list2 = ['Hello World', 'Python', 'JAVA']
xx = [str.lower() for str in list2]
print(xx)
```

**DICTIONAY COMPREHENSION**

```python
dictionary = {'a':1, 'b':2, 'c':3, 'd':4}
x = dictionary.keys()
print(x)

y = dictionary.values()
print(y)

z = dictionary.items()
print(z)
```

```python
NewDictionary = {k:v*2 for (k,v) in dictionary.items()}
print(NewDictionary)
```

**Dictionary comprehension as Lambda function**


```python
feh = {'temp1':10, 'temp2':20, 'temp3':30, 'temp4':40}
cel = list(map(lambda x: (float(5)/9)*(x-32), feh.values() ))
cel_dictionary = dict( zip(feh.keys(), cel))
print(cel_dictionary)
```

```python
cel2_dictionary = {k:(float(5)/9)*(v-32) for (k,v) in feh.items()}
print(cel2_dictionary)
```

```python
dictionary = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5, 'f':6}
new_dictionary = {k:('even No' if v%2 == 0 else 'odd No') for (k,v) in dictionary.items()}
print(new_dictionary)
```

**Nested dictionary comprehension**

```python
dictionary = {'one':{'a':10}, 'two':{'b':20}}
for (external_key, external_value) in dictionary.items():
  for (internal_key, internal_value) in external_value.items():
    external_value.update({internal_key: float(internal_value) + 10.3})
dictionary.update({external_key: external_value})
print(dictionary)
```

**Set comprehension**

```python
input = [1, 2, 2, 2, 3, 3, 3, 4, 5, 5, 6, 6, 7, 7, 7, 8]

output = set()
for var in input:
  if var % 2 == 0:
    output.add(var)
print(output)

output2 = {var for var in input if var % 2 == 0}
print(output2)
```

**Generator Comprehension**

```python
input = [1, 2, 3, 4, 4, 5, 6, 7, 7, 7]
output = (var for var in input if var % 2 == 0)
print(output)
print("output values using generator comprehension:", end='')
for var in output:
  print(var, end='')
```
