**Algoritmo de Ordenamiento - Quicksort**

```python
def quicksort(lista):
  # caso base
  if len(lista) <= 1:
    return lista
  
  # caso recursivo
  pivote = lista.pop()

  lista1, lista2 = [], []

  for e in lista:
    if e <= pivote:
      lista1.append(e)
    else:
      lista2.append(e)
  
  lista1 = quicksort(lista1)
  lista2 = quicksort(lista2)

  return lista1 + [pivote] + lista2

lista = [20, 19, 5, 6, 15, -4, 0, 18, 7, 3]

print(lista)
print(quicksort(lista))
```

```python
def binaria(lista, x):
  
  if len(lista) <= 0:
    return "Número no encontrado"

  m = lista[len(lista) // 2] # divide y se queda con la parte entera.

  if m == x:
    return "Número encontrado"
  else:
    if x < m: # el número que estamos buscando es menor a la mitad
      return binaria(lista[:len(lista) // 2], x)       # lista Izquierda
    else:
      return binaria(lista[(len(lista) //2) + 1 :], x) # lista Derecha

print(binaria(arreglo1, x))
```

**Algoritmo de Busqueda Hash**

Son funciones que crean, a partir de una dato de entrada, una salida finita que normalmente es una cadena de longitud fija. Esta cadena representa un resumen de la información de entrada.

Se debe construir una tabla **hash**

Se trata de una estructura de datos que asocia claves con valores. este tipo de estructura permite realizar búsquedas de orden constante.

Funcionan al transformar un valor o contenido con una función hash. Así se puede obtener un número que identifica la posición en la que se colocará el

```python
def convCad(cad):
  salida = ''

  for l in cad:        # Obtenemos el valor ascii de la cadena que queremos guardar
    salida += str(ord(l))
  
  return int(salida)
  

def hashM(cad, m):

  i = convCad(cad)  # convierto la cadena en un numero

  return int(m * (i * 0.00000000001 % 1))
```

```python
def agregar(cad, ht, m): # agrega en la tabla hash- recibe la cadena y la tabla
  res = hashM(cad, m)    # Tamaño de la tabla
  ht[res].append(cad)    # en la hash table, en la posicion que devuelve la pos. hash

def buscar(cad, ht, m):
  h = hashM(cad, m)

  for i in ht[h]: # elemento en la posicion hash de la tabla
    if i == cad:
      return True
  
  return False

# --------------------------------
m = 19
ht = [[] for i in range(m)] # listas vacias dentro de la lista
agregar('pizarron', ht, m)
agregar('plumon', ht, m)
agregar('borrador', ht, m)
agregar('impresora', ht, m)
agregar('pluma', ht, m)
agregar('cuaderno', ht, m)

print(ht)

print(buscar('pluma', ht, m))
print(buscar('libro', ht, m))
```

**GRAFICAS O GRAFOS**

```python
# Clase para definir un vértice
class Vertice:
  # constructor
  def __init__(self, i):
    self.id = i
    self.visitado = False
    self.nivel = -1
    self.vecinos = []

  # Agregar vecinos a un vertice
  def AgregarVecino(self, v): # v = vecino
    if v not in self.vecinos:
      self.vecinos.append(v)

# Clase para definir la gráfica
class Grafica:
  def __init__(self):
    self.vertices = {} # Diccionario, parecido a la lista pero no trabaja con indices

  def AgregaVertice(self, v):# chequea primero si no está
    if v not in self.vertices:
      self.vertices[v] = Vertice(v) # llama a la clase vertice
  
  def AgregarArista(self, a, b): # recibe los vértice de una arista
    if a in self.vertices and b in self.vertices:
      # defino que se puede ir en las dos direcciones de b-a y a-b 
      self.vertices[a].AgregarVecino(b) # dentro de la clase Vertice, la funcion
      self.vertices[b].AgregarVecino(a)


# --------------------------------
def main():

  g = Grafica()

  l = [0, 1, 2, 3, 4, 5, 6] # vertices
  for v in l:
    g.AgregaVertice(v) # agrego todos los vertices
  
  l = [2, 0, 0, 6, 6, 3, 0, 5, 6, 5, 0, 1, 6, 4, 1, 4] # Aristas
  for i in range(0, len(l) -1, 2): # recorriendo de 2 en dos, para la arista formada por dos vertices
    g.AgregarArista(l[i], l[i+1])
  
  for v in g.vertices:
    print(v, g.vertices[v].vecinos)
# -------------------------------------------

main()

```

**ITERATORS AND ITERABLES**

```python
class MyRange:

  def __init__(self, start, end):
    self.value = start
    self.end = end
  
  def __iter__(self): # Iterator
    return self

  def __next__(self):
    if self.value >= self.end:
      raise StopIteration
    current = self.value
    self.value += 1
    return current

nums = MyRange(1, 10)

for num in nums:
  print(num)

```

**Class**

```python
class Sentence:
  def __init__(self, sentence):  # Inicia la clase
    self.sentence = sentence
    self.index = 0 # recuerda donde está
    self.words = self.sentence.split() # parte la oración en una lista, con palabras espaciadas

  def __iter__(self): # 
    return self

  def __next__(self):
    if self.index >= len(self.words):
      raise StopIteration
    index = self.index
    self.index += 1
    return self.words[index]


my_sentence = Sentence('this is a test')

for word in my_sentence:
  print(word)
```

**Generator**

```python
def Sentence22(sentence):
  for word in sentence.split():
    yield word

my_sentence22 = Sentence22('this is a test')

for word in my_sentence22:
  print(word)
```
