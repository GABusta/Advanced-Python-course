```python
class teeShirt:
  def __init__(self, color, inventory_count):
    self.color = color
    self.inventory_count = inventory_count
  
  def __str__(self):
    return f"the {self.color} of the Tee shirts have {self.inventory_count} available"
  
  def __add__(self, other_self):
    result = self.inventory_count + other_self.inventory_count
    return result

black_shirt = teeShirt("black", 14)
white_shirt = teeShirt("white", 10)

print("color is: ", black_shirt.color)
print("inventory count is: ", black_shirt.inventory_count)
print("color is: ", white_shirt.color)
print("inventory count is: ", white_shirt.inventory_count)
print("-----------------------")
print(black_shirt)
# para poder operar con lo que guarda cada caja
print("-----------------------")
print(black_shirt + white_shirt)
```

**initialization and construction**

```python
class X:
  def __new__(cls):
    print("statement in the new magic method")
    inst = object.__new__(cls)
    return inst
  
  def __init__(self):
    print("this is the constructor function")
    self.name = "John"

# ------------
obj1 = X()
```

**Unary operators**

```python
# --- Positive
class X:
  def __init__(self, value):
    self._value = value
  
  def __pos__(self):
    print("pos magic method is called")
    return (+self._value)

# -------------------
obj1 = X(3)
print(+obj1)
```

```python
# --- Negative
class X:
  def __init__(self, value):
    self._value = value
  
  def __neg__(self):
    print("pos magic method is called")
    return (-self._value)

# -------------------
obj1 = X(3)
print(-obj1)
```

```python
# --- Inversion
class X:
  def __init__(self, value):
    self._value = value
  
  def __invert__(self):
    return self._value[::-1]
  def __str__(self):
    return self._value

# -------------------
obj1 = X('hello world')
inv_obj = ~obj1
print(inv_obj)
```

**Operators**

```python
class X:
  def __init__(self, x):
    self.x = x
  def __mul__(self, other):
    return self.x * other.x
# ---------
obj1 = X(4)
obj2 = X(3)
print(obj1 * obj2)
```

```python
class X:
  def __init__(self, x):
    self.x = x
  def __add__(self, other):
    return self.x + other.x
# ---------
obj1 = X(4)
obj2 = X(3)
print(obj1 + obj2)
```

**augmented assigment operator**

```python
class X:
  def __init__(self, x):
    self.x = x
  def __iadd__(self, other):
    self.x = self.x + other.x
    return self.x
  def __isub__(self, other):
    self.x = self.x - other.x
    return self.x

# ------------------------
obj1 = X(4)
obj2 = X(5)
obj1 += obj2
print(obj1)

obj3 = X(4)
obj4 = X(5)
obj3 -= obj4
print(obj3)
```
