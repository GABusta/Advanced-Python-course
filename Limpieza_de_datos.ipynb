**Carga de datos a través de la función Open**

```python
data2 = open('Customer Churn Model.txt','r')
```

```python
cols = data2.readline().strip().split(',')
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
  main_dict[col] = []
```

```python
for line in data2:
  values = line.strip().split(',')
  for i in range(len(cols)):
    main_dict[cols[i]].append(values[i])
  counter += 1

print('El data set tiene %d filas y %d columnas'%(counter, n_cols))
```

* ***Simulación de Monte Carlo***

  * Generamos 2 numeros aleatorios uniformes"x" e "y", entre 0 y 1

  * Calcularemos r*r = x*x + y*y
    * si el valor es menor que 1, dentro del circulo, sino fuera

  * Calculamos el número total de veces que están dentro del círculo y lo dividimos entre el número total de intentos para obtener una aproximación de la probabilidad de caer dentro del círculo.

  * Usamos dicha probabilidad para aproximar el valor de Pi.

  * Repetimos el experimento un número suficiente de veces, para obtener diferentes aproximaciones de Pi.

  * Calculamos el promedio de los experimentos anteriores para dar un valor final de Pi.

```python
pi_avg = 0
pi_value_list = []
n = 10000  # numero de puntos
nExp = 100    # numero de experimentos para los puntos
for i in range(nExp):
  value = 0
  x = np.random.uniform(0, 1, n).tolist()
  y = np.random.uniform(0, 1, n).tolist()
  for j in range(n):
    z = np.sqrt(x[j]*x[j] + y[j]*y[j])
    if z <= 1:
      value += 1
  float_value = float(value)
  pi_value = float_value * 4 / n
  pi_value_list.append(pi_value)
  pi_avg += pi_value

pi = pi_avg / nExp
print(pi)
plt.plot(pi_value_list)
```
