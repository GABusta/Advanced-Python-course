**Insertion - Singly linked list**

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
  
class LinkedList:
  def __init__(self):
    self.head = None

  def printing(self):
    printvalue = self.head
    while printvalue is not None:
      print(printvalue.data)
      printvalue = printvalue.next
  
  def beg(self, data4): # inserto un nuevo elemento al Inicio de la lista
    new = Node(data4)
    new.next = self.head
    self.head = new
  
  def end(self, data5): # inserto un nuevo elemento al final de la lista
    newest = Node(data5)
    if self.head is None:
      self.head = newest
      return
    lastNode = self.head
    while (lastNode.next):
      lastNode = lastNode.next
    lastNode.next = newest
  
  def between(self, node, datapart):
    newofall = Node(datapart)
    newofall.next = node.next
    node.next = newofall



x = LinkedList()
x.head = Node("Bustamante") # because "head" initiate in None
data2 = Node("Gustavo")
data3 = Node("Programa")
# --- conections ---
x.head.next = data2
data2.next = data3
# --- Print ---
# x.printing()

# --- Inserto una nueva estructura a traves de "beg" ---
x.beg("tutorials")
# x.printing()

# --- Inserto una nueva estructura a traves de "beg" ---
x.end("nada")
# x.printing()

# --- Inserto una nueva estructura a traves de "beg" ---
# agraga donde apuntaba "x"
x.between(x.head.next,"mmmm")
x.printing()
```

**Deletion - singly linked list**

```python
class Node:
  def __init__(self, data=None):
    self.data = data
    self.next = None


class LinkedList:
  def __init__(self):
    self.head = None

  def printing(self):
    printvalue = self.head
    while printvalue is not None:
      print(printvalue.data)
      printvalue = printvalue.next
  
  def delNode(self, deldata):
    new = self.head
    if new is not None:
      if new.data == deldata:
        self.head = new.next
        new = None
        return
        
    while new is not None:
      if new.data == deldata:
        break
      prev = new
      new = new.next

    if new == None:
      return
    
    prev.next = new.next
    new = None





x = LinkedList()
x.head = Node("Bustamante")
data2 = Node("gustavo")
data3 = Node("andres")
data4 = Node("programa")

x.head.next= data2
data2.next = data3
data3.next = data4

x.printing()
print("-----")

# --- deleting ---
x.delNode("andres")
x.printing()
```
