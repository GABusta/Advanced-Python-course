**New file**

```python
cgpa = {
    'Name':['Taylor', 'Ann', 'Christine'],
    'CGPA': [3.1, 2.2, 2.95]
}
df_cgpa = pd.DataFrame(cgpa)

favSports = {
    'Name':['Taylor', 'Ann', 'Christine'],
    'Fav Sports':['Basketball', 'Football', 'Tennis']
}
df_favSports = pd.DataFrame(favSports)

# merge two dataframes, en base a una columna en comun
merged_df = pd.merge(df_cgpa, df_favSports, on='Name')
print(merged_df)
```

new column 

```python
merged_df['Rank'] = merged_df['CGPA'].rank(ascending=1)
print(merged_df)
```

```python
def comment(x):
  if x>=2.8:
    return 'Above average'
  else:
    return 'Below average'

merged_df['Comment'] = merged_df['CGPA'].apply(comment)
print(merged_df)
```

```python
boolean = merged_df['Comment'].duplicated(keep=False)
new_df = merged_df[boolean]
print(new_df)
print('-----------')
cantidad = new_df['Comment'].count()
print(cantidad)
```
